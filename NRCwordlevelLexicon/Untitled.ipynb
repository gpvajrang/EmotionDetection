```python
from collections import OrderedDict, defaultdict, Counter
import pandas as pd
import csv
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('tweetdata.csv')
df.drop(columns =['id', 'Unnamed: 0', 'score'],axis=1, inplace=True)

```

```python
df.head()
```

```python
tweets = df['tweet']
```

```python
wordList = defaultdict(list)
emotionList = defaultdict(list)
with open('NRC-Emotion-Lexicon-Wordlevel-v0.92.txt', 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    headerRows = [i for i in range(0, 46)]
    for row in headerRows:
        next(reader)
    for word, emotion, present in reader:
        if int(present) == 1:
            #print(word)
            wordList[word].append(emotion)
            emotionList[emotion].append(word)
```

```python
tt = TweetTokenizer()
```

```python
def generate_emotion_count(string, tokenizer):
    emoCount = Counter()
    for token in tt.tokenize(string):
        token = token.lower()
        emoCount += Counter(wordList[token])
    return emoCount
```

```python
emotionCounts = [generate_emotion_count(tweet, tt) for tweet in tweets]
```

```python
emotion_df = pd.DataFrame(emotionCounts, index=tweets.index)
emotion_df = emotion_df.fillna(0)
```

```python
emotion_df.head()
```

```python
convert_dict = {'joy': int, 
                'positive': int,
                'trust':int,
                'anticipation':int,
                'anger':int,
                'negative':int,
                'sadness':int,
                'surprise':int,
                'fear':int,
                'disgust':int}
```

```python
emotion_df=emotion_df.astype(convert_dict)
```
